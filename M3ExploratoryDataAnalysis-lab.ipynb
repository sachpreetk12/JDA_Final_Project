## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
plt.figure(figsize=(8, 6))  # Set the size of the plot

df['ConvertedComp'].plot(kind='box')

plt.title('Distribution of Salary Compensation')
plt.ylabel('Annual USD Salary')

plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
Q1 = df['ConvertedComp'].quantile(0.25)  # First quartile (25th percentile)
Q3 = df['ConvertedComp'].quantile(0.75)  # Third quartile (75th percentile)

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

print("Interquartile Range (IQR) for 'ConvertedComp' column:", IQR)
```

Find out the upper and lower bounds.


```python
# your code goes here
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
# Identify outliers
outliers = df[(df['ConvertedComp'] < lower_bound) | (df['ConvertedComp'] > upper_bound)]

# Count the number of outliers
num_outliers = len(outliers)

print("Number of outliers in 'ConvertedComp' column:", num_outliers)
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
# Create a new DataFrame without outliers
df_no_outliers = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]

# Display the new DataFrame without outliers
print("New DataFrame without outliers in 'ConvertedComp' column created")
```

```python
df_no_outliers['ConvertedComp'].median()
```

```python
df_no_outliers['ConvertedComp'].mean()
```
