```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt

QUERY = """
SELECT Respondent, ConvertedComp
FROM master
"""
df_ConvertedComp = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(10, 6))  # Set the size of the figure (width, height)
df_ConvertedComp['ConvertedComp'].plot(kind = 'hist', color = 'orange')
plt.title('Distribution of Compensation in USD')
plt.xlabel('Annual Salaries')
plt.ylabel('Number of Respondents')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Respondent, Age
FROM master
"""
df_age = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(10, 6))  # Set the size of the figure (width, height)
df_age['Age'].plot(kind = 'box')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Number of Respondents')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
df_age_WWH = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(10, 6))  # Set the size of the figure (width, height)
df_age_WWH.plot(kind = 'scatter', x = 'Age', y = 'WorkWeekHrs', color = 'brown')
plt.title('Scatter plot for Age and Work Week Hours')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""
df_AWC = pd.read_sql_query(QUERY,conn)
size = df_AWC['Age']
df_AWC.plot(kind = 'scatter', x = 'WorkWeekHrs', y = 'CodeRevHrs',s = size* 10 ,alpha=0.5, color = 'purple', figsize=(10, 8))
plt.title('Scatter plot for Work Week and CodeRevHrs Hours with Age')
plt.xlabel('Work Week Hrs')
plt.grid(True)
plt.legend(['Age'], loc = 'upper right')
plt.ylabel('CodeRevHrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY = """
SELECT * 
FROM DatabaseDesireNextYear
"""
df_db = pd.read_sql_query(QUERY,conn)

# group respondents
group = df_db.groupby('DatabaseDesireNextYear', axis=0).count()

group.sort_values(["Respondent"], ascending=False, inplace=True)

df_top5 = group.head()
df_top5 
```

```python
df_top5['Respondent'].plot(kind='pie', figsize=(5, 6), autopct='%1.1f%%',startangle=90)
                     
plt.title('Top 5 Databases that Respondents Wish to Learn Next Year')
plt.axis('equal') # Sets the pie chart to look like a circle.

plt.show()
```

```python
QUERY = """
SELECT * FROM DevType
"""
df_dev = pd.read_sql_query(QUERY,conn)
df_dev = df_dev['DevType'].value_counts()
df_dev
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT CodeRevHrs, WorkWeekHrs, Age
FROM master
"""
df_stacked = pd.read_sql_query(QUERY,conn)

df_stacked = df_stacked[(df_stacked['Age'] >= 30) & (df_stacked['Age'] <= 35)]
median_workweek = df_stacked.groupby('Age')['WorkWeekHrs'].median()
median_coderev = df_stacked.groupby('Age')['CodeRevHrs'].median()
median_coderev
```

```python
fig, ax = plt.subplots()

# Plotting the median WorkWeekHrs
ax.bar(median_workweek.index, median_workweek, label='Median WorkWeekHrs', color = "#024b7a", alpha = 0.5)

# Plotting the median CodeRevHrs on top of WorkWeekHrs
ax.bar(median_coderev.index, median_coderev, bottom=median_workweek, label='Median CodeRevHrs', color = "#44a5c2", alpha = 0.5)

ax.set_xlabel('Age Group')
ax.set_ylabel('Hours')
ax.set_title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
ax.legend()
plt.xticks(median_workweek.index)  # Set x-axis ticks to be the age values

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT ConvertedComp, Age
FROM master
"""
df_line = pd.read_sql_query(QUERY,conn)

df_line = df_line[(df_line['Age'] >= 45) & (df_line['Age'] <= 60)]
median_compensation = df_line.groupby('Age')['ConvertedComp'].median()

plt.figure(figsize=(10, 6))  # Set the figure size

plt.plot(median_compensation.index, median_compensation, marker='o', color='#F9A729')

plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp for Ages 45-60')
plt.grid(True)
plt.xticks(median_compensation.index)  # Set x-axis ticks to be the age values

plt.show()

```
